```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sys

plt.style.use("../matplotlib_style/standard_plot_style.mplstyle")
```

```python
xs = np.linspace(0, 100, 10000)

def f(x, freq = 1):
    return np.cos(freq * x * 2 * np.pi)


fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (10, 8))

y1      = f(xs, 0.1)
y2      = f(xs, 0.15)
y3      = f(xs, 0.20)

ax0.plot(xs,  1 + 0.5 * y1, label = "1 Hz")
ax0.plot(xs,  0 + 0.5 * y2, label = "1.15 Hz")
ax0.plot(xs, -1 + 0.5 * y3, label = "1.3 Hz")

ax0.set(
    title = "Multiplexed signal",
    ylabel = "Amplitude"
)
ax1.set(
    xlabel = "Time [s]",

)

ax1.plot(xs, y1 + y2 + y3)

ax2 = ax1.inset_axes([0.6, 0.85, 0.2, 0.2])
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title("Fourier", fontsize = 12)
ax2.set_xlabel("frequency", fontsize = 10)
ax1.set_ylim(-5, 5)

y_fft = np.fft.rfft(y1 + y2 + y3)
ax2.plot(np.abs(y_fft)[:35])
```
