```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sys

plt.style.use("/mnt/c/Users/johan/OneDrive/Skrivebord/Master_Thesis/code/matplotlib_style/margin_figure.mplstyle")

import qutip

%matplotlib inline

save_figures = True
save_path    = "/mnt/c/Users/johan/OneDrive/Skrivebord/Master_Thesis/Figs/Theory/"

sys.path.append("../utils")

from matplotlib.colors import LinearSegmentedColormap
cmap_0 = LinearSegmentedColormap.from_list("mycmap", ["white", "C0"])
cmap_1 = LinearSegmentedColormap.from_list("mycmap", ["white", "C1"])
```

```python
from qutip import QFunc

lims = (-6, 6)

xvec = yvec = np.linspace(*lims, 201)

qfunc = QFunc(xvec, yvec)

fig, ax = plt.subplots(figsize = (4, 13), nrows = 3, sharey = True)

vacuum = qutip.ket2dm(qutip.basis(50, 0))
ax[0].imshow(qfunc(vacuum), extent = [*lims, *lims], origin = "lower", cmap = cmap_0)
ax[0].set(title = "Vacuum State", ylabel = r"$Im(\alpha)$")

fock_3 = qutip.ket2dm(qutip.basis(50, 3))
ax[1].imshow(qfunc(fock_3), extent = [*lims, *lims], origin = "lower", cmap = cmap_0)
ax[1].set(title = r"Fock State with $n = 3$", ylabel = r"$Im(\alpha)$")


coherent_alpha_2_2 = qutip.ket2dm(qutip.coherent(50, 2. + 2.j))
ax[2].imshow(qfunc(coherent_alpha_2_2), extent = [*lims, *lims], origin = "lower", cmap = cmap_0)
ax[2].set(title = r"Coherent State $\alpha = 2 + 2i$", xlabel = r"$Re(\alpha)$", ylabel = r"$Im(\alpha)$")


fig.tight_layout()

if save_figures:
    fig.savefig(save_path + "Q_functions.pdf", bbox_inches = "tight")
```
