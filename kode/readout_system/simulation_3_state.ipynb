```python

# Define Hamilton with given frequency
def get_Hamiltonian(omega):
    # Define the resonator given a pulse frequency
    readout_pulse = comp.ResonatorProbePulse(
        omega       = omega,
        duration    = (0, 50),
        rise_time   = 2,
        fall_time   = 2,
        amplitude   = 1 / 10
    )

    # Get pulse and the Unitary transformation opposite to the pulse
    pulse_on_resonator = readout_pulse.output()
    
    def Hamiltonian(t):
        constant_H = H_with_interaction

        # Define the pules and hamiltonian interacting with the resonator
        total_pulse = pulse_on_resonator[0](t) + pulse_on_resonator[1](t) 
        H_reso      = 2j *  total_pulse * (resonator.a_dagger() - resonator.a())
        H_reso_hilbert_space = np.kron(H_reso.todense(), transmon_identity)

        return constant_H + H_reso_hilbert_space

    return Hamiltonian


def master_equation(t, y, hamiltonian, U, Udot_inner = 0):
    # Get the hamiltonian and the unitary transformation
    H = hamiltonian(t)
    U = U(t)

    H_rotating = U @ H @ U.conj().T + 1j * U @ Udot_inner @ U.conj().T

    rho = y.reshape((system.dimensions[0] * system.dimensions[1], system.dimensions[0] * system.dimensions[1]))

    # Calculate the derivative of the density matrix
    d_rho = -1j * (H_rotating @ rho - rho @ H_rotating)

    return d_rho.flatten()
```

```python
# Define the first excited state of the resonator
resonator_g = np.zeros(resonator_states, dtype = np.complex128)
resonator_g[0] = 1

# Define ground state of transmon
transmon_0 = np.array((1, 0, 0), dtype = np.complex128)
transmon_1 = np.array((0, 1, 0), dtype = np.complex128)
transmon_2 = np.array((0, 0, 1), dtype = np.complex128)

# Define initial density matrix
initial_state_0 = np.kron(resonator_g, transmon_0)
initial_density_matrix_0 = np.outer(initial_state_0, initial_state_0.conj())

initial_state_1 = np.kron(resonator_g, transmon_1)
initial_density_matrix_1 = np.outer(initial_state_1, initial_state_1.conj())

initial_state_2 = np.kron(resonator_g, transmon_2)
initial_density_matrix_2 = np.outer(initial_state_2, initial_state_2.conj())
```

```python
from numba import prange
from tqdm import tqdm
from scipy.integrate import solve_ivp

# Define the frequencies to test
omegas_to_test = 2 * np.pi * np.linspace(5.92, 6.12, 7)

# Duration of Simulation
duration = (0, 50)


def simulate(omegas, initial_state):
    results = []

    for i in tqdm(range(len(omegas_to_test))):
        omega = omegas_to_test[i]
        
        H = get_Hamiltonian(omega)

        U, Udot_inner = get_rotating_frame(omega)
        
        sol = solve_ivp(master_equation, duration, initial_state.flatten(), args = (H, U, Udot_inner), method = 'RK45')

        results.append(sol)

    return results

print("Simulating state 0")
results_0 = simulate(omegas_to_test, initial_density_matrix_0)

print("Simulating state 1")
results_1 = simulate(omegas_to_test, initial_density_matrix_1)

print("Simulating state 2")
results_2 = simulate(omegas_to_test, initial_density_matrix_2)
```

```python
from scipy.special import factorial

def Q_function(density_matrix, n_cutoff = resonator_states, resolution = 20):
    xs = ys = np.linspace(-n_cutoff, n_cutoff, resolution)
    xs, ys = np.meshgrid(xs, ys)

    # Calculate alphas
    alphas = xs + 1j * ys
    alphas = alphas.flatten()

    # Noramlization
    normalization = np.exp(- alphas * alphas.conj() / 2)

    # Basis change
    ns = np.expand_dims(np.arange(n_cutoff), 1)
    P  = normalization * np.power(alphas, ns) / np.sqrt(factorial(ns))

    return np.diag(P.conj().T @ density_matrix @ P).reshape((resolution, resolution))
```

```python
def plot_Q(time, frequency_choice = 0, resolution = 20):
    sol_0 = results_0[frequency_choice].y.reshape(system.hilbert_size, system.hilbert_size, -1)
    sol_1 = results_1[frequency_choice].y.reshape(system.hilbert_size, system.hilbert_size, -1)
    sol_2 = results_2[frequency_choice].y.reshape(system.hilbert_size, system.hilbert_size, -1)

    # Get the reduced density matrix
    time_idx_0 = np.argmin(np.abs(results_0[frequency_choice].t - time))
    time_idx_1 = np.argmin(np.abs(results_1[frequency_choice].t - time))
    time_idx_2 = np.argmin(np.abs(results_2[frequency_choice].t - time))

    reso_0  = system.partial_trace(sol_0[:, :, time_idx_0], 1)
    reso_1  = system.partial_trace(sol_1[:, :, time_idx_1], 1)
    reso_2  = system.partial_trace(sol_2[:, :, time_idx_2], 1)

    # Get the Q function
    Q_0 = Q_function(reso_0, resolution=resolution)
    Q_1 = Q_function(reso_1, resolution=resolution)
    Q_2 = Q_function(reso_2, resolution=resolution)

    # Plot the Q function
    plt.imshow(abs(Q_0), cmap = "Blues", alpha = .5)
    plt.imshow(abs(Q_1), cmap = "Reds", alpha = .5)
    plt.imshow(abs(Q_2), cmap = "Greens", alpha = .5)

    plt.hlines(resolution / 2, 0, resolution, linestyles = "dashed", color = "black")
    plt.vlines(resolution / 2, 0, resolution, linestyles = "dashed", color = "black")

from ipywidgets import interact
interact(plot_Q, time = (0, 50), frequency_choice = (0, len(omegas_to_test) - 1, 1))
```

```python
dims = np.prod(system.dimensions)
def get_expectation_occupation(results):
    rhos = [ds.y.reshape(system.hilbert_size, system.hilbert_size, -1) for ds in results]
    last_rho = [rho[:, :, -1] for rho in rhos]
    last_rho_reso = [system.partial_trace(rho, trace_over = 1) for rho in last_rho]
    last_exp_n    = [np.trace(rho @ resonator.n_matrix()) for rho in last_rho_reso]

    return last_exp_n

exp_n_0 = get_expectation_occupation(results_0)
exp_n_1 = get_expectation_occupation(results_1)
exp_n_2 = get_expectation_occupation(results_2)
```
