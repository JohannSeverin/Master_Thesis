```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = lambda phi, alpha: - np.cos(phi + np.pi) + alpha * phi ** 2 / 2

xs = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

ys_low  = H(xs, 0.5)
ys_med  = H(xs, 1)
ys_high = H(xs, 2)
```

```python
plt.plot(xs, ys_med,    label = r"$\alpha = 1$")
plt.plot(xs, ys_low,    label = r"$\alpha = \frac{1}{2}$")
plt.plot(xs, ys_high,   label = r"$\alpha = 2$")

plt.legend()
```
