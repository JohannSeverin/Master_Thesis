```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define variables 
C1 = 100e-15     # fF
C2 = 100e-15     # fF
Cc = 100e-15    

C1c = 1e-15
C2c = 1e-15
C12 = 0.02e-15

w1 = w2 = 4e9 # GHz

delta =  1e9 # GHz ??? 

# Calculate interpolated variables

eta     = C1c * C2c / (C12 * Cc)



def find_g(wc):
    sigma_inv   = (1 / (w1 + wc) + 1 / (w2 + wc)) / 2   
    delta_inv   = (1 / (w1 - wc) + 1 / (w2 - wc)) / 2
    bracket     = delta_inv * wc * eta / 2 - sigma_inv * wc * eta / 2  + eta + 1 
    factor      = C12 / np.sqrt(C1 * C2) * np.sqrt(w1 * w2)
    return 1 / 2 * bracket * factor
    
```

```python
wcs = np.linspace(4.1e9, 7.0e9, 1000)

plt.plot(wcs, find_g(wcs))
plt.hlines(0, 4.1e9, 7.0e9, linestyles='dashed', colors='black')
```
