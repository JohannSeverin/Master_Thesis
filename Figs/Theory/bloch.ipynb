```python
b = qutip.Bloch()
b.vector_color = ["C0", "C1", "C2", "C3"]

b.make_sphere()

up = [0, 0, 1]
x  = [1, 0, 0]

test = [- 1/2, 1/2, 1 / np.sqrt(2)]
test_in_z = [0, 0, 1 / np.sqrt(2)]
test_in_x_y = [-1 / np.sqrt(2), 1 / np.sqrt(2), 0]

# b.add_vectors(up)
# b.add_vectors(x)
b.add_vectors(test)

b.add_arc(up, test, "k")
b.add_arc(x, test_in_x_y, "k")
b.add_line([0,0,0], test_in_x_y, "k--")
b.add_line([0,0,0], [1, 0, 0], "k--")
b.add_line([0,0,0], [0, 0, 1], "k--")
b.add_line([0,0,0], [-1/2, 1/2, 0], "C0-", alpha = 1.0)

b.add_annotation([0, 1/8, 1/3], r"$\theta$")
b.add_annotation([1/4, 1/4, 0], r"$\phi$")

b.render()
b.show()

b.save("bloch_sphere.pdf")
```
